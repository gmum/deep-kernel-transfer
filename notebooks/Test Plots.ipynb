```python
def plot_for_n(data,n=0):
    idx = np.argsort(data["x"][n],axis=0).reshape(-1)
    fig, (ax, ax2) = plt.subplots(1,2, figsize=(8*2,3*2))
    ax.plot(data["x"][n][idx], data["y"][n][idx, 0], linestyle="--", color="green")
    ax.plot(data["x"][n][idx], data["y"][n][idx, 0], linestyle='', marker='o', color="green")
    ax.plot(data["x"][n][idx], data["mean"][n][idx, 0])
    ax.fill_between(data["x"][n][idx].reshape(-1), data["lower"][n][idx, 0], data["upper"][n][idx, 0], alpha=0.2)

    ax2.plot(data["x"][n][idx], data["y"][n][idx, 1], linestyle="--", color="green")
    ax2.plot(data["x"][n][idx], data["y"][n][idx, 1], linestyle='', marker='o', color="green")
    ax2.plot(data["x"][n][idx], data["mean"][n][idx, 1])
    ax2.fill_between(data["x"][n][idx].reshape(-1), data["lower"][n][idx, 1], data["upper"][n][idx, 1], alpha=0.2)
    plt.show()
```
